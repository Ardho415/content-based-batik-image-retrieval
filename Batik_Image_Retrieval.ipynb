```python
import numpy as np
```

```python
def LBP(img):
  h,w = img.shape
  hasil = np.zeros((h,w), dtype=np.uint8)
  for i in range(1, h-1):
    for j in range(1, w-1):
      c = img[i,j]
      nilai = 0

      if img[i-1,j-1] >= c:
        nilai += 1
      if img[i-1,j] >= c:
        nilai += 2
      if img[i-1,j+1] >= c:
        nilai += 4
      if img[i,j+1] >= c:
        nilai += 8
      if img[i+1,j+1] >= c:
        nilai += 16
      if img[i+1,j] >= c:
        nilai += 32
      if img[i+1,j-1] >= c:
        nilai += 64
      if img[i,j-1] >= c:
        nilai += 128

      hasil[i,j] = nilai
  return hasil
```

```python
lbp_query = LBP(query)
print("LBP Query")
cv2_imshow(lbp_query)
```

```python
def Hamming(img1, img2):
  distance = 0
  for i in range(img1.shape[0]):
    for j in range(img1.shape[1]):
      xor = img1[i,j] ^ img2[i,j]
      distance += bin(xor).count('1')
  return distance
```
